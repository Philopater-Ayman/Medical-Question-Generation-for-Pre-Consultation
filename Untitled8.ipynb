```python
from datasets import load_dataset, DatasetDict
import random

# Load the dataset from Hugging Face Hub
print("Loading dataset 'Ahmed-Selem/Shifaa_Arabic_Medical_Consultations'...")
try:
    dataset = load_dataset("Ahmed-Selem/Shifaa_Arabic_Medical_Consultations")
    print("Dataset loaded successfully.")
    print("\nDataset structure:")
    print(dataset)
    print("\nExample data entry:")
    if 'train' in dataset:
        print(dataset['train'][0])
    else:
        first_split = list(dataset.keys())[0]
        print(dataset[first_split][0])

except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Please check the dataset name and your internet connection.")
    raise e

# --- Data Preprocessing ---
def format_instruction_data(example):
    """
    Formats a single example from the dataset into an instruction-response string.
    Uses 'Question Title' and 'Question' as input context to generate questions.
    """
    prompt = f"عنوان السؤال: {example.get('Question Title', '')}\n" \
             f"تفاصيل السؤال: {example.get('Question', '')}\n" \
             f"بناءً على هذه المعلومات الأولية، ما هي الأسئلة الإضافية التي يمكن طرحها للحصول على مزيد من التفاصيل قبل الاستشارة؟"

    formatted_text = f"{prompt}\n### الأسئلة المقترحة:\n"

    return {"text": formatted_text}

print("\nStarting dataset formatting...")
if 'train' in dataset:
    processed_dataset = dataset.map(format_instruction_data, remove_columns=dataset['train'].column_names)
    print("Dataset 'train' split formatted successfully.")
else:
    first_split_name = list(dataset.keys())[0]
    processed_dataset = dataset.map(format_instruction_data, remove_columns=dataset[first_split_name].column_names)
    print(f"Dataset '{first_split_name}' split formatted successfully.")

print("\nExample of formatted data entry:")
if 'train' in processed_dataset:
     print(processed_dataset['train'][0]['text'])
else:
    first_split = list(processed_dataset.keys())[0]
    print(processed_dataset[first_split][0]['text'])


# --- Dataset Splitting ---
print("\nStarting dataset splitting...")

if isinstance(processed_dataset, DatasetDict) and 'train' in processed_dataset:
    print("\nDataset has a 'train' split after formatting. Creating an 80/20 train/test split from it.")
    split_dataset = processed_dataset['train'].train_test_split(test_size=0.2, seed=42)
    train_dataset = split_dataset['train']
    eval_dataset = split_dataset['test']
    print("Created 'train' and 'test' splits with 80/20 ratio from the original 'train' split.")

elif not isinstance(processed_dataset, DatasetDict) and processed_dataset is not None and len(processed_dataset) > 0:
    print("\nDataset is a single Dataset object after formatting. Creating an 80/20 train/test split from it.")
    split_dataset = processed_dataset.train_test_split(test_size=0.2, seed=42)
    train_dataset = split_dataset['train']
    eval_dataset = split_dataset['test']
    print("Created 'train' and 'test' splits with 80/20 ratio.")

else:
    print("Error: The processed dataset is empty or in an unexpected format after formatting. Cannot perform splitting.")
    train_dataset = None
    eval_dataset = None


# Store the split datasets in a dictionary for easier access, if splitting was successful
if train_dataset is not None:
    processed_dataset_dict = DatasetDict({
        'train': train_dataset,
        'eval': eval_dataset
    })
    print("\nDataset splitting complete.")
    print(f"Training dataset size: {len(processed_dataset_dict['train'])} samples")
    if processed_dataset_dict['eval'] is not None:
        print(f"Evaluation dataset size: {len(processed_dataset_dict['eval'])} samples")
    else:
        print("No evaluation dataset available after splitting.")

else:
    print("\nDataset splitting skipped due to empty or unhandled dataset structure.")
    processed_dataset_dict = None


print("\nStep 2 completed.")
print("The 'processed_dataset_dict' variable now holds the split datasets.")
```

```python
import torch
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer
from transformers import TrainingArguments
import os
import gc # Garbage collection module

print("Starting Step 4: Fine-tuning google/gemma-3-1b-it using QLoRA...")

lora_config = LoraConfig(
    r=16,                      # LoRA attention dimension (rank). Common values are 8, 16, 32, 64.
    lora_alpha=16,             # Alpha parameter for LoRA scaling. Often set equal to r.
    lora_dropout=0.05,         # Dropout probability for LoRA layers.
    bias="none",               # Type of bias to add. "none" is common for QLoRA.
    task_type="CAUSAL_LM",     # Task type for causal language modeling.
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ]
)

print("\nApplying LoRA configuration to the model...")
model = get_peft_model(model, lora_config)
print("LoRA configuration applied.")
model.print_trainable_parameters()

train_subset_size = 500
eval_subset_size = 50

print(f"\nSelecting a subset of {train_subset_size} samples for training and {eval_subset_size} for evaluation...")

# Ensure the dataset dictionary and splits exist
if processed_dataset_dict is not None and 'train' in processed_dataset_dict:
    # Select a random sample from the training dataset
    if len(processed_dataset_dict['train']) > train_subset_size:
        train_subset = processed_dataset_dict['train'].shuffle(seed=42).select(range(train_subset_size))
        print(f"Selected {len(train_subset)} samples for training subset.")
    else:
        train_subset = processed_dataset_dict['train']
        print(f"Training dataset size ({len(train_subset)}) is less than or equal to requested subset size. Using entire training dataset.")

    # Select a random sample from the evaluation dataset if it exists
    eval_subset = None # Set to None initially
    if 'eval' in processed_dataset_dict and processed_dataset_dict['eval'] is not None:
        if len(processed_dataset_dict['eval']) > eval_subset_size:
            eval_subset = processed_dataset_dict['eval'].shuffle(seed=42).select(range(eval_subset_size))
            print(f"Selected {len(eval_subset)} samples for evaluation subset.")
        else:
            eval_subset = processed_dataset_dict['eval']
            print(f"Evaluation dataset size ({len(eval_subset)}) is less than or equal to requested subset size. Using entire evaluation dataset.")
    else:
        print("No evaluation dataset available or found. Skipping evaluation subset selection.")

else:
    print("Error: processed_dataset_dict is not available or does not contain a 'train' split. Cannot proceed with fine-tuning.")
    train_subset = None
    eval_subset = None

output_dir = "./gemma-3b-finetuned"
training_arguments = TrainingArguments(
    output_dir=output_dir,               
    per_device_train_batch_size=1,       
    gradient_accumulation_steps=8,         
    optim="paged_adamw_8bit",              
    learning_rate=2e-4,                  
    lr_scheduler_type="cosine",           
    save_steps=100,                        
    logging_steps=20,                     
    num_train_epochs=1,                    
    max_steps=-1,                       
    warmup_ratio=0.03,                     
    fp16=False,                           
    bf16=True,                            
    group_by_length=True,                 
    dataloader_num_workers=0,             
    report_to="none",                      # Reporting tool (e.g., "tensorboard", "wandb") - "none" for simplicity
    push_to_hub=False,                     # Whether to push the model to the Hugging Face Hub
)

if train_subset is not None:
    print("\nInitializing SFTTrainer...")
    trainer = SFTTrainer(
        model=model,                         # The PEFT model
        train_dataset=train_subset,          # The training dataset subset
        eval_dataset=eval_subset,            # The evaluation dataset subset (if available)
        peft_config=lora_config,             # The LoRA configuration
        # Removed tokenizer=tokenizer as it caused a TypeError
        args=training_arguments,             # The training arguments
    )

    # --- Start Training ---
    print("\nStarting training...")
    trainer.train()
    print("Training finished.")

    # --- Save the Fine-tuned Adapters ---
    # Only the LoRA adapters are saved, not the entire base model.
    print(f"\nSaving fine-tuned adapters to {output_dir}/final_checkpoint...")
    trainer.save_model(f"{output_dir}/final_checkpoint")
    print("Fine-tuned adapters saved.")

    # --- Clear Model from Memory ---
    # This is crucial to free up GPU memory for the next model.
    print("\nClearing model from memory...")
    del model
    del trainer
    torch.cuda.empty_cache() # Clear CUDA cache
    gc.collect()             # Run garbage collection
    print("Model cleared from memory.")

else:
    print("Skipping fine-tuning as the training dataset subset is not available.")

print("\nStep 4 completed.")
print("Fine-tuning for google/gemma-3-1b-it is finished (or skipped).")
```

```python
# @title Step 5: Load Gemma Adapters and Perform Inference

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel, PeftConfig
import os
import gc # Garbage collection module
from datasets import DatasetDict # Import DatasetDict to check type
import random # Import random for shuffling

print("Starting Step 5: Loading Gemma Adapters and Performing Inference...")

# --- Debugging Check ---
print("\nChecking for processed_dataset_dict availability...")
if 'processed_dataset_dict' in locals() and processed_dataset_dict is not None:
    print("processed_dataset_dict is found in local scope.")
    if isinstance(processed_dataset_dict, DatasetDict):
        print("processed_dataset_dict is a DatasetDict.")
        if 'eval' in processed_dataset_dict and processed_dataset_dict['eval'] is not None:
            print("processed_dataset_dict contains an 'eval' split.")
            print(f"Evaluation dataset size: {len(processed_dataset_dict['eval'])}")
        else:
            print("processed_dataset_dict does NOT contain an 'eval' split or it is None.")
    else:
        print("processed_dataset_dict is NOT a DatasetDict.")
else:
    print("processed_dataset_dict is NOT found in local scope or is None.")
# --- End Debugging Check ---

# --- Configuration for 4-bit Quantization ---
# Use the same quantization configuration as used for fine-tuning.
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=False,
)

# --- Load the Base Model ---
model_id = "google/gemma-3-1b-it"
print(f"\nLoading base model: {model_id} with 4-bit quantization...")

try:
    base_model = AutoModelForCausalLM.from_pretrained(
        model_id,
        quantization_config=bnb_config,
        device_map="auto",
        token=os.environ.get("HF_TOKEN")
    )
    print(f"Base model {model_id} loaded successfully with 4-bit quantization.")
    print(f"Base model device: {base_model.device}")

except Exception as e:
    print(f"Error loading base model {model_id}: {e}")
    print("Please check the model ID, your internet connection, and available resources.")
    raise e

# --- Load the Tokenizer ---
print(f"\nLoading tokenizer for base model: {model_id}...")
try:
    tokenizer = AutoTokenizer.from_pretrained(
        model_id,
        token=os.environ.get("HF_TOKEN")
    )
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    print(f"Tokenizer for {model_id} loaded successfully.")

except Exception as e:
    print(f"Error loading tokenizer for {model_id}: {e}")
    print("Please check the model ID and your internet connection.")
    raise e


# --- Load the Fine-tuned Adapters ---
adapter_path = "./gemma-3b-finetuned/final_checkpoint"
print(f"\nLoading fine-tuned adapters from: {adapter_path}...")

try:
    model = PeftModel.from_pretrained(base_model, adapter_path)
    print("Fine-tuned adapters loaded and merged with the base model.")

except Exception as e:
    print(f"Error loading PEFT adapters from {adapter_path}: {e}")
    print("Please ensure the adapter path is correct and the adapters were saved successfully in Step 4.")
    raise e

inference_sample_size = 5

print(f"\nSelecting {inference_sample_size} samples from the evaluation dataset for inference...")

inference_dataset = None
# Check if processed_dataset_dict and its 'eval' split are available
if 'processed_dataset_dict' in locals() and processed_dataset_dict is not None and 'eval' in processed_dataset_dict and processed_dataset_dict['eval'] is not None:
    if len(processed_dataset_dict['eval']) > inference_sample_size:
        # Select a random sample from the evaluation dataset
        inference_dataset = processed_dataset_dict['eval'].shuffle(seed=42).select(range(inference_sample_size))
        print(f"Selected {len(inference_dataset)} samples for inference.")
    elif len(processed_dataset_dict['eval']) > 0:
        inference_dataset = processed_dataset_dict['eval']
        print(f"Evaluation dataset size ({len(inference_dataset)}) is less than or equal to requested sample size. Using entire evaluation dataset.")
    else:
        print("Error: Evaluation dataset is empty. Cannot perform inference.")

else:
    print("Error: processed_dataset_dict is not available or does not contain an 'eval' split. Cannot perform inference.")


# --- Perform Inference (Actual Generation) ---
if inference_dataset is not None:
    print("\nGenerating questions using the fine-tuned Gemma model...")

    generated_questions_gemma = []
    evaluation_prompts = [] # Store the original prompts for evaluation

    for i, example in enumerate(inference_dataset):
        # Use the formatted text from the dataset as the input prompt
        prompt = example['text']
        evaluation_prompts.append(prompt) # Store the prompt

        # Prepare input for the model
        input_ids = tokenizer(prompt, return_tensors="pt").to(model.device)

        # Generate questions (This will attempt actual generation)
        with torch.no_grad(): # Disable gradient calculation for inference
            output_tokens = model.generate(
                **input_ids,
                max_new_tokens=100, # Limit the length of generated questions
                num_return_sequences=1,
                no_repeat_ngram_size=2, # Avoid repeating n-grams
                do_sample=True,         # Enable sampling for more diverse questions
                top_k=50,               # Sample from top K tokens
                top_p=0.95,             # Sample using nucleus sampling
                temperature=0.7,        # Control randomness
                pad_token_id=tokenizer.pad_token_id # Use the defined pad token
            )

        # Decode the generated tokens
        generated_text = tokenizer.decode(output_tokens[0], skip_special_tokens=True)

        # Extract only the generated questions part after the prompt separator
        separator = "### الأسئلة المقترحة:\n"
        if separator in generated_text:
            generated_questions = generated_text.split(separator, 1)[1].strip()
        else:
            # If the model didn't generate after the separator, the generated_questions will be empty.
            generated_questions = "" # Ensure it's an empty string if separator not found or nothing follows

        generated_questions_gemma.append(generated_questions)

        print(f"\n--- Sample {i+1} ---")
        print(f"Original Prompt:\n{prompt}")
        print(f"\nGenerated Questions (Gemma):\n{generated_questions}") # Print the actual generated questions (might be empty)

    print("\nInference with fine-tuned Gemma model completed.")
    print("Generated questions and original prompts are stored in 'gemma_generated_questions' and 'evaluation_prompts'.")

else:
    print("\nSkipping inference as evaluation dataset subset is not available.")
    generated_questions_gemma = None
    evaluation_prompts = None


# --- Clear Model from Memory ---
print("\nClearing model from memory...")
# Check if model and base_model exist before deleting
if 'model' in locals() and model is not None:
    del model
if 'base_model' in locals() and base_model is not None:
    del base_model
if 'tokenizer' in locals() and tokenizer is not None:
    del tokenizer # Clear tokenizer as well
torch.cuda.empty_cache()
gc.collect()
print("Model cleared from memory.")

print("\nStep 5 completed.")
```

```python
# @title Step 9: Gemini API Setup and Evaluation

import google.generativeai as genai
import os
import pandas as pd # Using pandas to store evaluation results
import gc # Garbage collection module
import re # Import regex for parsing evaluation results

print("Starting Step 9: Gemini API Setup and Evaluation...")
try:
    # Attempt to get the API key from Colab Secrets
    from google.colab import userdata
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
    if GOOGLE_API_KEY is None:
         print("Warning: 'GOOGLE_API_KEY' not found in Colab Secrets.")
         print("Attempting to use hardcoded API key as a fallback.")
         GOOGLE_API_KEY = "AIzaSyB45g_hzbtqPxggTbB5v9R-jp9IK9JHhW8"
         if GOOGLE_API_KEY == "AIzaSyB45g_hzbtqPxggTbB5v9R-jp9IK9JHhW8": # Check if the placeholder was replaced
              print("Warning: Using hardcoded API key. This is not secure and should be replaced with a Colab Secret.")
         else:
              # This case should ideally not be reached if using the provided key
              raise ValueError("Gemini API key not found in Secrets or hardcoded.")


    genai.configure(api_key=GOOGLE_API_KEY)
    print("Gemini API configured successfully.")

except Exception as e:
    print(f"Error configuring Gemini API: {e}")
    print("Please ensure your API key is correct and accessible.")
    # Stop execution if API setup fails
    raise e # Re-raise the exception

# --- Initialize Gemini Model (Judge Model) ---

judge_model_name = "gemini-1.5-flash-latest"
print(f"\nInitializing Gemini judge model: {judge_model_name}...")
try:
    judge_model = genai.GenerativeModel(judge_model_name)
    print(f"Gemini judge model {judge_model_name} initialized.")
except Exception as e:
    print(f"Error initializing Gemini judge model {judge_model_name}: {e}")
    print("Please check the model name and your API key permissions.")
    raise e # Re-raise the exception


# --- Evaluation Function ---
# This function will prompt Gemini to evaluate the generated questions.
def evaluate_generated_questions(original_context, generated_questions, ground_truth_answer):
    # Construct the prompt for Gemini
    prompt_template = """
    أنت مقيّم لأسئلة المتابعة الطبية. سيتم تزويدك بسياق طبي أولي (عنوان السؤال وتفاصيله) ومجموعة من أسئلة المتابعة المقترحة التي تم إنشاؤها بواسطة نموذج لغوي.
    مهمتك هي تقييم جودة أسئلة المتابعة المقترحة على مقياس مستمر من 0 إلى 100. تعتبر المجموعة جيدة إذا كانت تتضمن جميع المعلومات الرئيسية المتعلقة بالسياق الأولي وتساعد في جمع معلومات إضافية مهمة للطبيب، وإذا كانت جميع الأسئلة المطروحة ذات صلة ومفهومة.
    **السياق الطبي الأولي (السؤال الأساسي والتفاصيل):**
    {original_context}

    **أسئلة المتابعة المقترحة :**
    {generated_questions}

    **الأسئلة الصحيحة (الحقيقة الأرضية - هذا هو رد الطبيب الأصلي للمريض، وليس بالضرورة أسئلة المتابعة التي كان يجب طرحها قبل الاستشارة. استخدمه كسياق إضافي للحكم على مدى ملاءمة الأسئلة المولدة للسياق الأولي):**
    {ground_truth}

    **التقييم:**
    النتيجة (الدرجة): (على مقياس من 0 إلى 100)
    التعليل (السبب): (شرح موجز لسبب النتيجة، مع الإشارة إلى الملاءمة والوضوح والفائدة بناءً على السياق الأولي. مثال: السؤال ذو صلة، لكنه يفتقر إلى بعض التفاصيل حول الأعراض الرئيسية التي يجب على الطبيب ملاحظتها. السؤال يحتاج إلى تعديل ليشمل الأعراض الرئيسية.)

    **فكر خطوة بخطوة:**
    1. هل السؤال الذي تم توليده يغطي الجوانب الرئيسية للسياق؟
    2. هل هناك أي معلومات غير صحيحة أو مضللة في السؤال؟
    3. هل يساعد السؤال الطبيب في جمع المعلومات الضرورية من المريض؟
    4. هل السؤال واضح وسهل الفهم؟
    """

    full_prompt = prompt_template.format(
        original_context=original_context,
        generated_questions=generated_questions,
        ground_truth=ground_truth_answer # Pass the original Answer as ground truth context
    )

    try:
        # Send the prompt to the Gemini model
        # Use generate_content with a reasonable timeout
        response = judge_model.generate_content(full_prompt, request_options={'timeout': 60}) # Add a timeout
        # Extract the text from the response
        evaluation_result = response.text
        return evaluation_result

    except Exception as e:
        print(f"Error during Gemini evaluation: {e}")
        return f"Evaluation failed: {e}"

# Helper function to parse score and reasoning from Gemini's text output
def parse_evaluation_result(evaluation_text):
    # Use regex to find the score (a number after "النتيجة (الدرجة):")
    score_match = re.search(r"النتيجة \(الدرجة\):\s*(\d+)", evaluation_text)
    # Use regex to find the reasoning (text after "التعليل (السبب):")
    # re.DOTALL allows '.' to match newline characters
    reasoning_match = re.search(r"التعليل \(السبب\):(.*)", evaluation_text, re.DOTALL)

    score = int(score_match.group(1)) if score_match else None
    reasoning = reasoning_match.group(1).strip() if reasoning_match else "Reasoning not found."

    return score, reasoning


# --- Perform Evaluation ---
if 'evaluation_prompts' in locals() and evaluation_prompts is not None and len(evaluation_prompts) > 0 and \
   'gemma_generated_questions' in locals() and gemma_generated_questions is not None and len(gemma_generated_questions) == len(evaluation_prompts) and \
   'silma_generated_questions' in locals() and silma_generated_questions is not None and len(silma_generated_questions) == len(evaluation_prompts) and \
   'processed_dataset_dict' in locals() and processed_dataset_dict is not None and 'eval' in processed_dataset_dict and processed_dataset_dict['eval'] is not None and \
   len(evaluation_prompts) <= len(processed_dataset_dict['eval']): # Ensure we don't exceed eval dataset size


    print(f"\nPerforming evaluation using Gemini for {len(evaluation_prompts)} samples...")

    evaluation_results = []
    eval_dataset_subset = processed_dataset_dict['eval'].select(range(len(evaluation_prompts)))


    for i in range(len(evaluation_prompts)):
        original_context = evaluation_prompts[i]
        gemma_questions = gemma_generated_questions[i]
        silma_questions = silma_generated_questions[i]
        # Get the corresponding original Answer from the eval dataset subset
        # Use .get() with a default to avoid KeyError if 'Answer' is missing
        ground_truth_answer = eval_dataset_subset[i].get('Answer', 'N/A - Answer field missing in dataset')


        print(f"\n--- Evaluating Sample {i+1} ---")
        # Print original context and generated questions for clarity
        print(f"Original Context:\n{original_context}")
        print(f"\nOriginal Doctor's Answer (used as Ground Truth Context):\n{ground_truth_answer}")
        print(f"\nGemma Generated Questions:\n{gemma_questions}")
        print(f"\nSILMA Generated Questions:\n{silma_questions}")


        # Evaluate Gemma's generated questions using the Gemini API
        print("\nEvaluating Gemma's questions with Gemini...")
        gemma_evaluation_raw = "Gemma did not generate any questions for this sample." # Default message
        gemma_score = None
        gemma_reasoning = "Evaluation skipped or failed."

        if gemma_questions and gemma_questions.strip(): # Check if the string is not empty or just whitespace
            gemma_evaluation_raw = evaluate_generated_questions(original_context, gemma_questions, ground_truth_answer)
            print("Gemma Raw Evaluation Result from Gemini:")
            print(gemma_evaluation_raw) # Print the raw output from Gemini
            gemma_score, gemma_reasoning = parse_evaluation_result(gemma_evaluation_raw)
            print(f"  Gemma Score: {gemma_score if gemma_score is not None else 'N/A'}")
            print(f"  Gemma Reasoning: {gemma_reasoning}")
        else:
            print(gemma_evaluation_raw) # Print the default message


        # Evaluate SILMA's generated questions using the Gemini API
        print("\nEvaluating SILMA's questions with Gemini...")
        silma_evaluation_raw = "SILMA did not generate any questions for this sample." # Default message
        silma_score = None
        silma_reasoning = "Evaluation skipped or failed."

        if silma_questions and silma_questions.strip(): # Check if the string is not empty or just whitespace
            silma_evaluation_raw = evaluate_generated_questions(original_context, silma_questions, ground_truth_answer)
            print("SILMA Raw Evaluation Result from Gemini:")
            print(silma_evaluation_raw) # Print the raw output from Gemini
            silma_score, silma_reasoning = parse_evaluation_result(silma_evaluation_raw)
            print(f"  SILMA Score: {silma_score if silma_score is not None else 'N/A'}")
            print(f"  SILMA Reasoning: {silma_reasoning}")
        else:
            print(silma_evaluation_raw) # Print the default message


        # Store results (can parse the text output later for structured analysis)
        evaluation_results.append({
            "sample_index": i + 1,
            "original_context": original_context,
            "ground_truth_answer": ground_truth_answer, # Store the ground truth answer used
            "gemma_generated_questions": gemma_questions,
            "silma_generated_questions": silma_questions,
            "gemma_evaluation_raw": gemma_evaluation_raw,
            "silma_evaluation_raw": silma_evaluation_raw,
            "gemma_score": gemma_score, # Store parsed score
            "gemma_reasoning": gemma_reasoning, # Store parsed reasoning
            "silma_score": silma_score, # Store parsed score
            "silma_reasoning": silma_reasoning # Store parsed reasoning
        })

    print("\nEvaluation using Gemini completed.")
    if evaluation_results:
        evaluation_df = pd.DataFrame(evaluation_results)
        print("\nEvaluation results stored in 'evaluation_df' DataFrame.")
    else:
        print("\nNo evaluation results to store.")

else:
    print("\nSkipping evaluation. Required variables (evaluation_prompts, generated_questions, processed_dataset_dict) are not available or lengths mismatch.")
    print("Please ensure Step 2, Step 7 (Gemma Inference), and Step 8 (SILMA Inference) were run successfully and generated outputs.")


print("\nStep 9 completed.")
```
